```python
import pandas as pd

df = pd.read_csv("dataset.csv")
df.head()  # Check the first 5 rows
```

```python
# Baseline Pricing Function
def baseline_price(prev_price, occupancy, capacity, alpha=5):
    if capacity == 0:
        return prev_price
    occ_rate = occupancy / capacity
    new_price = prev_price + alpha * occ_rate
    new_price = min(max(new_price, 5), 20)  # Limit price between $5 and $20
    return round(new_price, 2)
```

```python
# Apply model on first 10 rows for testing
prev_price = 10  # Starting price
prices = []

for i in range(10):  # Use full len(df) later if needed
    occ = df.loc[i, 'Occupancy']   # Capital O
    cap = df.loc[i, 'Capacity']    # Capital C
    price = baseline_price(prev_price, occ, cap)
    prices.append(price)
    prev_price = price  # Update price for next row

# Print generated prices
for i, p in enumerate(prices):
    print(f"Time Step {i+1}: ₹{p}")
```
